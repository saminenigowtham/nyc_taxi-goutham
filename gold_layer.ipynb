## creating database

```python
%sql
create schema taxi_nyc_catlog.gold;
```

```python
df_zone = spark.read.format('parquet')\
            .option('inferSchema', 'true')\
            .option('header', 'true')\
            .option('path','abfss://silver@taxinycdatalakegoutham.dfs.core.windows.net/trip_zone_lookup')\
            .load()
```

```python
df_zone.display()
```

```python
spark.sql("CREATE SCHEMA IF NOT EXISTS db_nyc_taxi_goutham.gold")

df_zone.write.format('delta')\
        .mode('overwrite')\
        .option('path','abfss://gold@taxinycdatalakegoutham.dfs.core.windows.net/trip_zone')\
        .saveAsTable('db_nyc_taxi_goutham.gold.trip_zone')
```

```python
%sql
select * from gold.trip_zone;
```
